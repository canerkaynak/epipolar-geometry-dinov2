```python
import concurrent.futures
import kagglehub

base_dir = '/content/drive/MyDrive/epipolar-geometry-dinov2/epipolar-geometry-dinov2/data/kitti_sample'

kagglehub.dataset_download('hocop1/kitti-odometry', path='poses/00.txt', output_dir=base_dir)
kagglehub.dataset_download('hocop1/kitti-odometry', path='sequences/00/calib.txt', output_dir=base_dir)
kagglehub.dataset_download('hocop1/kitti-odometry', path='sequences/00/times.txt', output_dir=base_dir)

def download_image(folder_name, file_index):
    file_name = str(file_index).zfill(6)
    kagglehub.dataset_download(
        'hocop1/kitti-odometry',
        path=f'sequences/00/{folder_name}/{file_name}.png',
        output_dir=base_dir
    )

with concurrent.futures.ThreadPoolExecutor(max_workers=10) as executor:
    for idx in range(150):
        executor.submit(download_image, 'image_2', idx)
        executor.submit(download_image, 'image_3', idx)
```
